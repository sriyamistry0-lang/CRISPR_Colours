```python
def diagnostic_scan(sample, library):
    print(f"--- ðŸ¥ Clinical Diagnostic Scan Starting ---")
    found_threats = []

    for gene_name, sequence in library.items():
        # Using 're' allows us to find the position easily
        match = re.search(sequence, sample)

        if match:
            start, end = match.span()
            print(f"âš ï¸ ALERT: {gene_name} DETECTED!")
            print(f"   Location: Position {start} to {end}")
            found_threats.append({"Gene": gene_name, "Start": start, "End": end})

    if not found_threats:
        print("âœ… CLEAR: No known resistance genes found in this sample.")

    return pd.DataFrame(found_threats)

# --- THE TEST DATA ---
# Let's say we have our library from before
resistance_library = {
    "blaTEM-1": "ATAAAATTCTTGAAGACGAAA",
    "mecA": "GGGATATAGTCAGTTCCAACT",
    "tetA": "GCTCATATTCTGCTCTCA"
}

# The 'Patient' DNA
mystery_dna = "CGTTAA" + "GGGATATAGTCAGTTCCAACT" + "TAGCCT"

# Run the engine
report = diagnostic_scan(mystery_dna, resistance_library)
```

```python
import random

# 1. GENERATE THE 'HUMAN HAYSTACK'
def generate_human_haystack(gene_to_hide):
    human_dna_chars = ['A', 'T', 'G', 'C']
    # Create 10,000 bases of random 'Human' DNA
    haystack = "".join(random.choices(human_dna_chars, k=10000))

    # Hide the Superbug gene at a random position
    insert_pos = random.randint(1000, 8000)
    infected_sample = haystack[:insert_pos] + gene_to_hide + haystack[insert_pos:]

    return infected_sample, insert_pos

# Our NDM-1 'Target'
ndm1_target = "ATGGAATTGCCCAATATTATGCACCCGGTCGCGAAGCCCTGACCGCGCC"

# Create the clinical sample
patient_swab, true_pos = generate_human_haystack(ndm1_target)

print(f"ðŸ“Š Metagenomic Sample Generated: {len(patient_swab)} bp")
print("Status: Patient is symptomatic. Running Diagnostic Terminal...")

# 2. RUN YOUR SCANNER
import re
match = re.search(ndm1_target, patient_swab)

if match:
    print("\n" + "="*40)
    print("ðŸš¨ CLINICAL ALERT: NDM-1 DETECTED")
    print(f"Detected at: Position {match.start()}")
    print(f"Actual Hidden Position: {true_pos}")
    print("="*40)
    print("âœ… VERIFICATION: Code successfully filtered human DNA.")
else:
    print("âŒ FAILED: The pathogen was missed in the human noise.")
```

```python
def calculate_gc(sequence):
    gc_count = sequence.count('G') + sequence.count('C')
    return (gc_count / len(sequence)) * 100

# Let's check the 'Human' part vs the 'Pathogen' part
human_segment = patient_swab[0:100]
pathogen_segment = patient_swab[6968:7017] # Where our match was found

print(f"ðŸ“Š Human GC-Content: {calculate_gc(human_segment):.2f}%")
print(f"ðŸ§¬ Pathogen GC-Content: {calculate_gc(pathogen_segment):.2f}%")

if abs(calculate_gc(human_segment) - calculate_gc(pathogen_segment)) > 5:
    print("\nðŸš¨ OBSERVATION: Significant GC-shift detected. This confirms 'Foreign' DNA is present.")
```
